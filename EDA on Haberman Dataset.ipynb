# EDA on Haberman Cancer Survival Dataset

### The dataset contains data supporting studies between 1958 and 1970 at University of Chicago on patients' survival after undergoing breast cancer surgery.

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from google.colab import files
```

```
haberman = files.upload()
```

```
haberman = pd.read_csv('haberman.csv')
```

```
haberman.shape
```

```
haberman.columns
```

```
haberman.head()
```

```
haberman['status'].value_counts()
```

First columns specifies the patients' age, second specifies year of surgery, third specifies number of tumors found and the last column shows the survival status- 1 being 5 years or more and 2 being less than 5 years

```
haberman.describe()
```

Age: The mean age of patients comes out as 62 years. 
50 % of patients are of age approx. 63.
Year: Patient with age 58 years is the youngest to undergo surgery.
75% patients had operation at age 66.

Nodes: More than 25 % had no nodes, averaging around 4 per person.

status: more than 50 % patients survived longer than 5 years.

```
haberman['nodes'].value_counts()
```

```
haberman['year'].value_counts()
```

```
#plotting a scatter plot
haberman.plot.scatter(x ='age', y = 'nodes')
plt.show()
```

```
sns.set_style('whitegrid')
sns.FacetGrid(haberman, hue = 'status', size = 8).map(plt.scatter, "age", "nodes").add_legend()
plt.show()
```

Most patients had no nodes.
Most patients with no nodes have for the longest times(> 5 years).

## Mean, Std-Dev and Variance

```
not_dead = haberman.loc[haberman['status'] == 1]
dead = haberman.loc[haberman['status'] == 2]
```

```
print("Patients surviving more than 5 years: ")
not_dead.describe()
```

```
print("Patients surviving no more than 5 years: ")
dead.describe()
```

## Percentiles, Median, Quantiles, IQR and MAD

```
#printing medians
print(np.median(not_dead['nodes']))
print(np.median(dead['nodes']))
```

```
#printing quantiles
print(np.percentile(not_dead['nodes'], np.arange(0, 100, 25)))
print(np.percentile(dead['nodes'], np.arange(0, 100, 25)))
```

```
#printing percentiles
print(np.percentile(not_dead['nodes'], 25))
print(np.percentile(dead['nodes'], 25))
print(np.percentile(not_dead['nodes'], 50))
print(np.percentile(dead['nodes'], 50))
print(np.percentile(not_dead['nodes'], 75))
print(np.percentile(dead['nodes'], 75))
```

```
from statsmodels import robust as rb
print(rb.mad(not_dead['nodes']))
print(rb.mad(dead['nodes']))
```

## Box plot and Whisker plots

```
sns.boxplot(x = 'status', y = 'nodes', data = haberman)
plt.show()
```

We observe least outlier values in status and nodes columns

```
sns.boxplot(x = 'status', y = 'age', data = haberman)
plt.show()
```

Patients survived with more than 5 years are aged between 40 and 60 years.

## Violin Plot

```
fig, axes = plt.subplots(1,3,figsize = (16,8))
for id, column in enumerate(list(haberman.columns)[:-1]):
  sns.violinplot(x = 'status', y = column, data = haberman, ax = axes[id])

plt.show()
```

- 80% patients have less than or equal to 5 lymph nodes and survived more than 5 years


## Contour Plot

```
sns.jointplot(x = 'age', y = 'year', data = haberman, kind = "kde")
plt.show()
```

- Most patients underwent surgery in ages [40,60]
